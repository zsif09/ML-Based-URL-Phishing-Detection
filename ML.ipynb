```python
#To avoid imbalence train set
seed = 13

from sklearn.model_selection import StratifiedShuffleSplit
split_data=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=seed)
for train_index,test_index in split_data.split(dataset,dataset['Label']):
    strat_train_set=dataset.loc[train_index]
    strat_test_set=dataset.loc[test_index]
```

```python
strat_test_set['Label'].value_counts(),strat_train_set['Label'].value_counts()
```

```python
dataset=strat_train_set.copy()
testdata=strat_test_set.copy()
dataset.shape,testdata.shape
```

```python
dataset_x=strat_train_set.drop(['Domain','Label'],axis=1)
dataset_y=strat_train_set["Label"].copy()
dataset_x.shape,dataset_y.shape
```

```python
test_x=strat_test_set.drop(['Domain','Label'],axis=1)
test_y=strat_test_set["Label"].copy()
```

```python
dataset_x.head()
```

```python
dataset_y.head()
```

```python
dataset_x, X_val, dataset_y, y_val = train_test_split(dataset_x, dataset_y, test_size=0.1, random_state=seed, stratify=dataset_y)
```

```python
pd.value_counts(dataset_y), pd.value_counts(y_val)
```

#Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc, confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import mean_squared_error

forest = RandomForestClassifier(n_estimators=300,random_state=42,max_features=.75)
history = forest.fit(dataset_x,dataset_y)

predicted_train_y= forest.predict(dataset_x)
mse = mean_squared_error(dataset_y, predicted_train_y)
print("Mean Squared Error: Train", mse)

print('--------------------------------------------')

predicted_y_val= forest.predict(X_val)
mse = mean_squared_error(y_val, predicted_y_val)
print("Mean Squared Error: Val", mse)

print('--------------------------------------------')

predicted_y= forest.predict(test_x)
mse = mean_squared_error(test_y, predicted_y)
print("Mean Squared Error: Test", mse)

print('-----------------TEST---------------------------')
# Convert the predicted probabilities to binary labels
predicted_labels = np.round(predicted_y).astype(int)

# Generate the classification report
report = classification_report(test_y, predicted_labels, digits = 4)
# Print the classification report
print(report)

print('-----------------VAL---------------------------')
# Convert the predicted probabilities to binary labels
predicted_labels = np.round(predicted_y_val).astype(int)

# Generate the classification report
report = classification_report(y_val, predicted_labels, digits = 4)
# Print the classification report
print(report)
```
